```python
#Converting the oject type to float type
data['HDI_Female'] = pd.to_numeric(data['HDI_Female'],errors='coerce')
data['HDI_Male'] = pd.to_numeric(data['HDI_Male'],errors='coerce')
data['GDI_Value'] = pd.to_numeric(data['GDI_Value'],errors='coerce')
data['Lif_Excep_Male'] = pd.to_numeric(data['Lif_Excep_Male'],errors='coerce')
data['Lif_Expec_Female'] = pd.to_numeric(data['Lif_Expec_Female'],errors='coerce')
data['Excep_Yrs_Schooling_Male'] = pd.to_numeric(data['Excep_Yrs_Schooling_Male'],errors='coerce')
data['Excep_Yrs_Schooling_Female'] = pd.to_numeric(data['Excep_Yrs_Schooling_Female'],errors='coerce')
```

```python
#Total Life Expectancy of Male
data['Total_lif_Male']=data['HDI_Male']
data['Total_lif_Male']=data['Total_lif_Male']*data['Lif_Excep_Male']
data.head()
```

```python
#Total Life Expectancy of Female
data['Total_lif_Female']=data['HDI_Female']
data['Total_lif_Female']=data['Total_lif_Female']*data['Lif_Expec_Female']
data.head()
```
